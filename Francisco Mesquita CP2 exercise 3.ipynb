```python
N=1000

for n in range(1,26):
    Zn=np.zeros(N)
    for j in range(N):
        for k in range (25):
            Zn[j]+=np.random.uniform()
    Zn+=-0.5*n
    const=np.sqrt(12)/np.sqrt(n)
    Zn=const*Zn
    
    print('N=',n)
    print("Mean=",np.mean(Zn))
    print("Variance=",np.var(Zn))
    plt.hist(Zn,bins=20)
    plt.show()
```
